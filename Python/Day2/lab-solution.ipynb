# Strings and Lists in python

1-	Fill an array of 5 elements from the user, Sort it in descending
and ascending orders then display the output

---

```python
el_list = []
for i in range(5):
    el = input("Enter elements ")
    el_list.append(el)
sorted_el_asc = sorted(el_list)
sorted_el_desc = sorted(el_list, reverse=True)
print(f"ASC: {sorted_el_asc}, DESC: {sorted_el_desc}")
```

2-Write a program that prints the number of times the string 'iti'
occurs in anystring.

---

```python
text = input("Enter any string: ")
count = 0

for i in range(len(text) - 2):
    if text[i:i+3] == "iti":
        count += 1

print(count)
```

3- Write a program that remove all vowels from the input word and
generate a brief version of it

---

```python
vowels = ['a', 'e', 'i', 'o', 'u']
word = input('Enter a word')
for char in word.lower():
    if char in vowels:
        word = word.replace(char, '')

print(word)
```

4- Write a program that counts up the number of vowels **[a, e, i, o,u]** contained in the string.

```python
vowels = ['a', 'e', 'i', 'o', 'u']
word = input('Enter a word')
count = 0
for char in word:
    if char in vowels:
        count += 1

print(count)
```

7-  Write a Python script to:
- Take a string input from the user.
- Convert the string to uppercase.
- Count the number of vowels in the string.
- Reverse the string.
- Check if the string is a **palindrome**
--  example: madam

```python
text = input("Enter a string: ")

upper_text = text.upper()
print("Uppercase:", upper_text)

vowels = ['a', 'e', 'i', 'o', 'u']
count = 0
for char in text:
    if char in vowels:
        count += 1

print(count)

reversed_text = text[::-1]
print("Reversed ", reversed_text)

if text == reversed_text:
    print("palindrome")
else:
    print("not palindrome")
```

# Sets in python

### 1- Create a set of the first five even numbers. Write a script to:

- Add a new number to the set.
- Remove a number from the set.
- Check if a specific number exists in the set.
- Perform a union and intersection with another set `{2, 4, 6, 8, 10}`.


```python
even_numbers = {2, 4, 6, 8, 10}

print("Original :", even_numbers)

even_numbers.add(12)
print(even_numbers)
```

```python

even_numbers.remove(4)
print(even_numbers)

number_to_check = 6
if number_to_check in even_numbers:
    print(f"{number_to_check} exists")
else:
    print(f"{number_to_check} does not exist")
```

```python
another_set = {2, 4, 6, 8, 10}

union_set = even_numbers.union(another_set)
print("Union:", union_set)

intersection_set = even_numbers.intersection(another_set)
print("Intersection:", intersection_set)
```
