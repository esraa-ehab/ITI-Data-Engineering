6) Find the Largest of Three Numbers
Take three numbers as input and print the largest one.

```python
nums = input("Enter numbers separated by commas: ")
nums_list = nums.split(",")
max = nums_list[0]
for i in nums_list:
    if i>max:
        max = i
print(f"maximum number is : {max}")
```

10) Prime Number Checker
Ask the user for a number and check if it's a prime number.

```python
import math

num = int(input("Enter number"))

if num <2:
    is_prime = False
else:
    is_prime = True
    for i in range(2, math.ceil(math.sqrt(num))+1):
        if num%i == 0 and num != i :
            is_prime = False
if is_prime: print("Prime")
else: print("Not Prime")
```

11- Write a program that asks the user for:

 - The total number of students enrolled in a course

 - The desired number of students per group

The program should then calculate and display the total number of groups formed. If the students cannot be divided evenly, one group may contain fewer members than specified.

**Example output (1):**

- How many students on the course? 8
- Desired group size? 4
- Number of groups formed: 2


**Example output (2):**

- How many students on the course? 11
- Desired group size? 3
- Number of groups formed: 4


> Hint: The integer division operator // may be helpful in solving this problem.

```python
import math
students = int(input("Enter number of students: "))
students_per_group = int(input("Enter number of students per group: "))
num_of_groups = math.ceil(students/students_per_group)
print(f" number of groups = {num_of_groups}")
```
